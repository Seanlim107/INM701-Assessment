```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data=pd.read_csv("stroke_prediction_dataset.csv")

# data.head()
data=data.drop(['Patient ID','Patient Name'],axis=1)

data["Gender"]=np.where(data["Gender"]=="Female",0,1)

data['Married']=np.where(data['Marital Status']=='Married',1,0)
data['Divorced']=np.where(data['Marital Status']=='Divorced',1,0)
dummies=pd.get_dummies(data['Work Type'],dtype=float)
data=data.join(dummies)
data=data.drop(['Marital Status','Work Type'],axis=1)

data['Residence Type']=np.where(data['Residence Type']=='Rural',0,1)

smoking_cond=[data['Smoking Status']=='Non-smoker',data['Smoking Status']=='Formerly Smoked',data['Smoking Status']=='Currently Smokes']
smoking_dec=[0,0.5,1]
data['Smoking Status']=np.select(smoking_cond,smoking_dec,default='unknown')


data.head()
# data["Marital Status"]=np.where(data["Marital Status"]==)
```
